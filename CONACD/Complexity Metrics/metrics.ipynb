```python
from numpy.random import seed
from numpy.random import rand
import numpy as np
import matplotlib.pyplot as plt
```

```python
centroide_1=[1, 1]
centroide_2=[1, 1]
centroide_3=[6, 6]
std_dev_1 = 1
std_dev_2 = 2
std_dev_3 = 1
n_points_1 = 500
n_points_2 = 10000
n_points_3 = 500
```

```python
# J = (m2-m1)²/(s1²+s2²)
# J1 = (1-5)²/(1²+1²)
# J1 = 16/2 = 8 (maximizar)

# J2D = (raiz(1²+5²)+raiz(1²+5²))²/2
# J2D = (r2-r50)²/2
# J2D = -5.66²/2
# J2D = 16
```

```python
def create_random_point_centroides(centroide, n_points, std_dev):
    
    num_centroide = len(centroide)
    point_lists = np.random.randn(num_centroide, n_points)
    
    for i in range(len(point_lists)):
        for j in range(len(point_lists[i])):
            point_lists[i][j] = point_lists[i][j] * std_dev + centroide[i]
    return point_lists
```

```python
point_lists1 = create_random_point_centroides(centroide = centroide_1, n_points = n_points_1, std_dev = std_dev_1)

point_lists2 = create_random_point_centroides(centroide = centroide_2, n_points = n_points_2, std_dev = std_dev_2)
point_lists3 = create_random_point_centroides(centroide = centroide_3, n_points = n_points_3, std_dev = std_dev_3)
```

```python
plt.plot(point_lists2[0],point_lists2[1], 'ro', point_lists1[0],point_lists1[1], 'bo', point_lists3[0],point_lists3[1], 'ro')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
